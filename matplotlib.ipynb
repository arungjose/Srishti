```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
y = [10,20,40,50]
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, color='green', linestyle='--', marker='o')
plt.title('My Plotter')
plt.xlabel('x-axis')
plt.ylabel('y-label')
```

```python
x = [1,2,3,5]
y1 = [1,4,5,8]
y2 = [2,5,8,9]
plt.plot(x, y1, label='Squares')
plt.plot(x, y2, label='Linear')
plt.legend()
plt.show()
```

```python
categories = ['A', 'B', 'C']
values = [5,6,7]
plt.bar(categories, values)
plt.show()
```

```python
data = [1,2,3,4,5,6,7,8,9]
plt.hist(data, bins=5)
plt.show()
```

```python
x = [1,2,3,4]
y = [10,20,25,30]
plt.scatter(x,y)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot([1,2,3], [1,2,3])
plt.subplot(1,2,2)
plt.plot([1,2,3], [3,2,1])
plt.show()
```

```python
plt.plot(x,y)
plt.savefig('plot.png')
```

```python
plt.plot(x, y)
plt.fill_between(x, y1=y, y2=0, alpha=0.2);
```

```python
x = ['Jan', 'Feb', 'March', 'Apr', 'May']
y1 = [1, 2, 3, 4, 5]
y2 = [2, 5, 9, 13, 18]
y3 = [2, 4, 6, 8, 10]
plt.plot(x, y1, x, y2, x, y3)
plt.xlabel('Month')
plt.ylabel('Sales');
```
